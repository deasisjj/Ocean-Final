```
import datetime
import glob
#dimensianlize data
t = np.array([], dtype=object)
d = np.array([], dtype=object)
for file_name in glob.iglob('./drive/My Drive/OceanDataFInals/2017-2019/*.csv'): #/drive/My Drive/OceanDataFInals/2014-2015/**/*.csv'
  #content/drive/My Drive/OceanDataFInals/2014-2015
  dat = np.genfromtxt(file_name, delimiter=',', skip_header = 1)
  #print(dat.shape)
  time = (dat[:,1].astype('datetime64[s]') - np.datetime64('2040-01-02T00:00:00Z'))/np.timedelta64(1, 's')
  time = time.astype('datetime64[s]')
  #print(time.shape)
  #print(prs)
  temp = dat[:,13]
  #print(tmp)
  t = np.concatenate([t, time])
  d = np.concatenate([d, temp])
  #plt.plot(time, temp)
  
u = np.hstack([t.reshape(-1,1), d.reshape(-1,1)])
#plt.plot(t, d)
u = np.array(sorted(u, key= lambda x: x[0]))
plt.plot(u[:,0], u[:,1])
plt.title('Temperature vs. Time')
plt.xlabel('Time in year-month')
plt.ylabel('Temperature in Celcius')

plt.show()
```
